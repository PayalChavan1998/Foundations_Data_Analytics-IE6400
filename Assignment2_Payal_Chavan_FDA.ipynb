### Que 2) In a poker hand consisting of 5 cards, find the probability of holding: a. All 4 aces b. 1 pair, Jacks or better (Jacks, Queens, Kings, or Aces).

```python
import math

# a)To find the probability of holding all 4 aces

# Total no. of aces in 52 cards is 4
# No. of ways to choose 4 aces from 4 aces = C(4,4)
all_4_aces = math.comb(4,4)

# No. of ways to choose fifth card from remaining 48 cards = C(48,1)
fifth_card = math.comb(48,1)

# Total no. of possible combinations = C(52,5)
total_hands = math.comb(52,5)

# Calculate the probability 
P_4_aces = (all_4_aces * fifth_card) / total_hands
print("a) Probability of holding all 4 aces:", round(P_4_aces, 8))


# b)To find 1 pair, Jacks or better

# Choosing 1 card from 4 options((Jacks, Queens, Kings, or Aces) = C(4,1)
choose_1_card = math.comb(4, 1)

# Choosing 2 cards to form pair from name card = C(4,2)
choose_2_cards = math.comb(4, 2)

# Choosing other 3 cards from remanining cards = C(48,3)
choose_3_cards = math.comb(48, 3)

# Total number of possible ways = C(52,5)
total_ways = math.comb(52, 5)

# Calculate the probability
P_Jacks_or_Better = (choose_1_card * choose_2_cards * choose_3_cards) / total_ways

print("b) Probability of holding 1 pair Jacks or better:", round(P_Jacks_or_Better, 4))
```

a) Combination formula = nCr = n!/(n-r)! * r!
Probability = (C(4, 4) * C(48, 1)) / C(52, 5) = 1.85 * 10^-5

b) Combination formula = nCr = n!/(n-r)! * r!
Probability = (C(4, 1) * C(4, 2) * C(48, 3)) / C(52, 5) = 0.1597

### Que 4) A teacher is creating an exam with 12 questions. Fortunately, the teacher has a question bank with 30 possible questions to use. Assuming the teacher chooses the questions at random and the order of the questions is irrelevant, how many different groups of questions could the teacher use to create the exam?

```python
import math

# Number of questions in the question bank
n = 30

# Number of questions to be chosen for the exam
r = 12

# Calculate the number of combinations
nCr = math.comb(n, r)

print(" The number of different groups of questions:", nCr)
```

Combination formula= nCr = n!/(n-r)! * r!
30C12 = 𝐶(30,12) = 30!/12!*(30−12)! = 86,493,225

### Que 5) An outdoor gear manufacturing company is testing their process for creating new backpacks. Each backpack has 2 straps that should be the same exact length; however, the process is not perfect. The manufacturing machine may create 3 different scenarios: A, straps exactly the same length; B, right strap longer than left; C, left strap longer than right. Suppose that P(A) = 0.945 and P(C) = 0.024. a. Find the probability that the right strap is longer, P(B) b. What is the probability that the left strap is not longer than the right? c. What is the probability of the machine creating straps that aren’t the same length?

```python
# Probability of scenario A - Straps exactly the same length
P_A = 0.945

# Probability of scenario C - left strap longer than right
P_C = 0.024

# Probability of scenario B - right strap longer than left
# P(A) + P(B) + P(C) = 1
P_B = 1 - P_A - P_C
print("a) Probability that the right strap is longer:", round(P_B, 3))

# Probability that left strap is not longer than right
P_not_longer = 1 - P_C
print("b) Probability that the left strap is not longer than the right:", round(P_not_longer, 3))

# Probability of straps not of the same length
P_not_same_length = 1 - P_A
print("c) Probability of the machine creating straps that aren’t the same length:", round(P_not_same_length, 3))
```

### Que 6) There is a lecture hall that hosts a class with 25 students. There are 9 coveted seats in the back row, where most students prefer to sit. Assuming all students come to class, how many different ways can the 9 seats in the back row be filled?

```python
import math

# Number of students in the class
n = 25

# Number of coveted seats in the back row
r = 9

# Calculate the number of permutations
nPr = math.perm(n, r)

print("The number of different ways to fill the 9 seats in the back row:", nPr)
```

Permutation formula = nPr = n!/(n-r)!
25P9 = P(25,9) = 25!/(25-9)! = 741,354,768,000

### Que 7) In a data analysis class of 210 students, 87 students commute by bus, 72 students live in on-campus housing, and 43 commute by bus AND live in on-campus housing. If one student is selected at random, find the probability that: a. The student commutes by bus OR lives in on-campus housing b. The student neither commutes by bus NOR lives in on-campus housing c. The student commutes by bus, but does not live in on-campus housing

```python
# Given data
total_students = 210
bus_students = 87    #P(A)
on_campus_students = 72  #P(B)
both_students = 43    #P(C)

# Probability of commute by bus OR lives in on-campus housing
# P(A OR B) = P(A) + P(B) - P(A AND B)
P_bus_or_on_campus_students = (bus_students + on_campus_students - both_students) / total_students

# Probability of commute neither by bus NOR lives in on-campus housing
# P(A NOR B) = 1 - P(A OR B)
P_neither_bus_nor_on_campus_students = 1 - P_bus_or_on_campus_students

# Probability of commute by bus but NO on-campus housing
# P(A and NOT B) = P(A) - P(A OR B)
P_bus_not_on_campus = (bus_students - both_students) /  total_students

print("a) The probability that the student commutes by bus OR lives in on-campus housing is", 
      round(P_bus_or_on_campus_students, 3))
print("b) The probability that the student neither commutes by bus NOR lives in on-campus housing is",
      round(P_neither_bus_nor_on_campus_students, 3))
print("c) The probability that the student commutes by bus, but does not live in on-campus housing is",
      round(P_bus_not_on_campus, 3))
```

### Que 8) Every day, many vehicles cross the border from Canada into Washington state. The probability that one of these vehicles has Washington license plates is 0.47; the probability that it contains Seattle residents is 0.21; the probability of it having Washington plates AND containing Seattle residents is 0.16. Calculate the probability that: a. A Seattle resident entering from Canada has Washington plates b. An entering vehicle with Washington plates contains a Seattle resident c. A vehicle entering does not have Washington plates or is not a Seattle resident (i.e. missing at least 1 of the traits)

```python
# Given data
P_Washington = 0.47    #P(A)
P_Seattle = 0.21       #P(B)
P_Washington_And_Seattle = 0.16   #P(C)

# Probability of Seattle resident has Washington plates
# P(A|B) = P(A AND B) / P(B)
P_Seattle_resident_Washington_plates = P_Washington_And_Seattle / P_Seattle

# Probability of Washington plates contains Seattle resident
# P(B|A) = P(A AND B) / P(A)
P_Washington_plates_Seattle_resident = P_Washington_And_Seattle / P_Washington

# Probability of no Washington plates or is not a Seattle resident
# P(not A OR not B) = 1 - P(A AND B)
P_not_Washington_plates_or_not_Seattle_resident = 1 - P_Washington_And_Seattle


print("a) The probability of Seattle resident entering from Canada has Washington plates is",
      round(P_Seattle_resident_Washington_plates, 3))
print("b) The probabiity of vehicle with Washington plates contains a Seattle resident is", 
      round(P_Washington_plates_Seattle_resident, 3))
print("c) The probability of vehicle entering does not have Washington plates or is not a Seattle resident is", 
      round(P_not_Washington_plates_or_not_Seattle_resident, 3))
```

### Que 9) Two professional darts players are competing in a match. The first, Xavier, has a 0.27 probability of hitting a bullseye. The second, Yvette, has a 0.31 probability of hitting a bullseye. The probability of both hitting a bullseye is 0.14. Find the probability that: a. Either Xavier or Yvette hit a bullseye b. Neither Xavier nor Yvette hit a bullseye

```python
# Given data
P_X = 0.27    #P(A)
P_Y = 0.31    #P(B)
P_X_AND_Y = 0.14   #P(C)

# Probability of Either Xavier or Yvette hit a bullseye
# P(A OR B) = P(A) + P(B) - P(A AND B)
P_X_OR_Y = P_X + P_Y - P_X_AND_Y

# Proability of Neither Xavier nor Yvette hit a bullseye
# p(A NOR B) = 1 - P(A OR B)
P_X_NOR_Y = 1 - P_X_OR_Y

print("a) The probability of Either Xavier or Yvette hit a bullseye is", round(P_X_OR_Y, 3))
print("b) The proability of Neither Xavier nor Yvette hit a bullseye is", round(P_X_NOR_Y, 3))
```
