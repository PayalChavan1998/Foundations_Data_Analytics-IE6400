## TASK 1: Computation of inter-event time

```python
# Define a custom function to compute inter-event times

import numpy as np

def inter_event_func(time_series, origin):
    events = np.where(time_series >= origin)[0]
    inter_event_func = np.diff(events)
    return inter_event_func
```

```python
# Demonstrate the function using the given array & a threshold of 5

time_series = np.array([10,1,6,4,3,2,7,0,8,8,2,0,7,7,1,0])
origin = 5
inter_event_func = inter_event_func(time_series, origin)
print(inter_event_func)
```

### Que 1: Create a table that shows the top 5 states by number of unique ports.

```python
# Find the number of port names based on states

top_states = border_crossings_data.groupby("State")["Port Code"].nunique().reset_index().sort_values("Port Code",ascending=False)

# Display table by top 5 States & No. of unique ports

top_states.iloc[:,:2].rename(columns={"Port Code": "Total no. of Ports"}).head()
```

### Que 2: Summarize the truck crossings in 4 years(2019-2022)

```python
# Copy the raw data
truck_crossings = border_crossings_data.copy(deep=True)

# Transform column "Date" to datetime type
truck_crossings["Date"] = pd.to_datetime(truck_crossings["Date"])

truck_crossings.info()
```

```python
# Filter the data for the last 4 years(2019-2022)
truck_crossings_data = truck_crossings[(truck_crossings["Date"].dt.year >= 2019) & (truck_crossings["Date"].dt.year <= 2022)]


# Filter the data for truck crossings
truck_crossings_data = truck_crossings_data[truck_crossings_data["Measure"] == "Trucks"]

# Calculating total monthly trucks
truck_crossings_data = truck_crossings_data.groupby([truck_crossings_data["Border"],truck_crossings_data["Date"].dt.year]).sum().reset_index()
truck_crossings_data = truck_crossings_data.rename(columns={"Date":"Year","Value":"TotalMonthlyTrucks"})

# Calculating monthly average trucks
truck_crossings_data["AverageMonthlyTrucks"]= truck_crossings_data["TotalMonthlyTrucks"] / 12

# Display the table
truck_crossings_data = truck_crossings_data[["Border","Year","AverageMonthlyTrucks","TotalMonthlyTrucks"]]
display(truck_crossings_data.sort_values(by="Year"))
```

### Que 3: Comparing the crossing types side-by-side for each port on each date.

```python
# Pivot the table
pivot_df = border_crossings_data.pivot_table(index=["Port Name","State","Border","Date"], columns="Measure", values="Value").reset_index()

# Fill missing values with 0
pivot_df.fillna(0)

pivot_df.columns.name = None
pivot_df.head()
```

### Que 4: Create a table showing the monthly totals by state, sorted chronologically first, then by greatest TotalVehicles.

```python
# Create a table with the following column names
pivot_df["TotalVehicles"] = pivot_df[[
    "Buses", "Personal Vehicles", "Trains", "Trucks"]].sum(axis=1)

# Grouping the data by State & Year
pivot_df = pivot_df.groupby(["State", "Date"], as_index=False)["TotalVehicles"].sum(
).sort_values(by=["Date", "TotalVehicles"], ascending=[True, False])
display(pivot_df)
```

### Que 5: Border crossings under Republican versus Democratic presidents. 

```python
presidential_parties = border_crossings_data.copy(deep=True)
presidential_parties["Date"] = pd.to_datetime(presidential_parties["Date"])

# Filtering data based on the range of the years
presidential_parties = presidential_parties[(presidential_parties["Measure"].isin(["Personal Vehicle Passengers", "Train Passengers", "Pedestrians", "Bus Passengers"]))]

range_condition = np.logical_or(np.logical_or(np.logical_and(presidential_parties["Date"].dt.year >= 1989, presidential_parties["Date"].dt.year <= 1992), np.logical_and(
    presidential_parties["Date"].dt.year >= 2001, presidential_parties["Date"].dt.year <= 2008)), np.logical_and(presidential_parties["Date"].dt.year >= 2016, presidential_parties["Date"].dt.year <= 2020))
#[1989,1990,1991,1992,2001,2002,2003,2004,2005,2006,2007,2008,2016,2017,2018,2019,2020]

presidential_parties["PresidentialParty"] = np.where(range_condition, "Republican", "Democratic")

border_presidential_parties = presidential_parties.groupby([presidential_parties["Date"].dt.year,"PresidentialParty"])["Value"].sum().reset_index()
border_presidential_parties = border_presidential_parties.rename(columns={"Value":"IndividualCrossings"})

display(border_presidential_parties)
```

### Que 6: Combine homicide data with border crossings data & also use the state abbreviations data

```python
# Converting Abbreviation to uppercase
state_abbreviations_data["Abbreviation"] = state_abbreviations_data["Abbreviation"].str.upper()

state_abbreviations_data = state_abbreviations_data.rename(columns={"Abbreviation":"STATE"})

merged_df = pd.merge(state_homicides_data, state_abbreviations_data, on="STATE")

merged_df = merged_df.groupby(['State','YEAR'],as_index=False)['DEATHS'].sum()
```

```python
merged_df = merged_df.rename(columns={'YEAR':'Year','DEATHS':'Deaths'})
```
